```python
user_usage_dict = pickle.load( open( "user_usage_dict.pickle", "rb" ) )
```

```python
for i,key in enumerate(user_usage_dict.keys()):
    if i >2:
        break
    else:
        print(key)
```

```python
ax = plt.subplot(111)
ax.bar(x-0.2, y, width=0.2, color='b', align='center')
ax.bar(x, z, width=0.2, color='g', align='center')
ax.bar(x+0.2, k, width=0.2, color='r', align='center')
ax.xaxis_date()

plt.show()
```

```python
plot_dict = defaultdict(defaultdict)
tier_types = [16.0,32.0,61.0,64.0,72.0,512.0]
for idx,userId in enumerate(user_usage_dict.keys()):
    if idx>5:
        break
    else:
        for mem in tier_types:
            if mem in user_usage_dict[userId].keys():
                plot_dict[userId][str(mem)] = user_usage_dict[userId][mem][-1]
            else:
                plot_dict[userId][str(mem)] = 0
```

```python
plt.figure(figsize=(48,12))
for idx,userId in enumerate(plot_dict.keys()):
    ax = plt.subplot(181+idx)
    plt.title('Normalized cpu usage vs memory for user{}'.format(userId[-3:]))
    ax.bar(list(plot_dict[userId].keys()), list(plot_dict[userId].values()), width=0.9, color=['black', 'red', 'green', 'blue', 'cyan'], align='center')

plt.show()
```
